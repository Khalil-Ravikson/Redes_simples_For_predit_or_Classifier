```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size = 0.25)
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape = (4,)),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 3, activation = 'softmax')
])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

```python
rede_neural.fit(X_treinamento, y_treinamento, batch_size = 15, epochs = 1000)
```

```python
rede_neural.evaluate(X_teste, y_teste)
```

## VALIDAÇÂO CRUZADA


```python
def criarRede():
  rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape = (4,)),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 3, activation = 'softmax')])
  rede_neural.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
  return rede_neural
```

```python
rede_neural = KerasClassifier(build_fn = criarRede, epochs = 250, batch_size = 10)
```

```python
resultados = cross_val_score(estimator = rede_neural, X = X_treinamento, y = y_treinamento, cv = 10, scoring = 'accuracy')
```
