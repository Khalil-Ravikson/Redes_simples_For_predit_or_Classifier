```python
import scikeras
import pandas as pd
import tensorflow as tf
import sklearn
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
```

# COLOCAR NO GOOGLE O ARQUIVO .CSV E ADCIONAR X E y :

```python
X = pd.read_csv('/content/entradas_breast.csv')
```

```python
y = pd.read_csv('/content/saidas_breast.csv')
```

```python
def rede_neural():
  k.clear_session()
  rede_neural= Sequential([
      tf.keras.layers.InputLayer(shape=(30,)),
      tf.keras.layers.Dense(units=16,activation='relu',kernel_initializer='random_uniform'),
      tf.keras.layers.Dropout(0.2),
      tf.keras.layers.Dense(units=16,activation='relu',kernel_initializer='random_uniform'),
      tf.keras.layers.Dropout(0.2),
      tf.keras.layers.Dense(units=1,activation='sigmoid'),])
  otimizador= tf.keras.optimizers.Adam(learning_rate=0.001,clipvalue=0.5)
  rede_neural.compile(optimizer=otimizador,loss='binary_crossentropy',metrics=['accuracy'])
  return rede_neural
```

```python
rede_neural = KerasClassifier(model=rede_neural,epochs=100,batch_size=10)
```

```python
resultados = cross_val_score(estimator=rede_neural,X=X,y=y,cv=10,scoring='accuracy',error_score='raise')
```

```python
resultados
```

```python
resultados.mean()
```

```python
resultados.std()
```

# TUNING

```python
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
```

```python
def rede_neural_tunning(optimizer, loss, kernel_initializer, activation, neurons):
  k.clear_session()
  rede_neural= Sequential([
      tf.keras.layers.InputLayer(shape=(30,)),
      tf.keras.layers.Dense(units=neurons,activation=activation,kernel_initializer=kernel_initializer),
      tf.keras.layers.Dropout(0.2),
      tf.keras.layers.Dense(units=neurons,activation=activation,kernel_initializer=kernel_initializer),
      tf.keras.layers.Dropout(0.2),
      tf.keras.layers.Dense(units=1,activation='sigmoid'),])
  rede_neural.compile(optimizer=optimizer,loss=loss,metrics=['accuracy'])
  return rede_neural
```

```python
rede_neural_Tunada= KerasClassifier(model=rede_neural_tunning)
```

```python
parametros = {
    'batch_size':[10,15,30],
    'epochs':[10,20,50],
    'model__optimizer':['adam'],
    'model__loss':['binary_crossentropy'],
    'model__kernel_initializer':['random_uniform'],
    'model__activation':['relu'],
    'model__neurons':[16,8,4]
}
```

```python
parametros = {
    'batch_size':[10,30],
    'epochs':[50,100],
    'model__optimizer':['adam','sgd'],
    'model__loss':['binary_crossentropy','hinge'],
    'model__kernel_initializer':['random_uniform','normal'],
    'model__activation':['relu','tanh'],
    'model__neurons':[16,8]
}
```

```python
grid_search = GridSearchCV(estimator=rede_neural_Tunada,param_grid=parametros,scoring='accuracy',cv=5)
```

```python
grid_search = grid_search.fit(X,y)
```

```python
print(grid_search)
```

```python
print(grid_search.best_params_)
```
