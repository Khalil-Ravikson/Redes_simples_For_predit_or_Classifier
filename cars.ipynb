```python
def criar_modelo():
    regressor = Sequential([
        tf.keras.layers.InputLayer(shape=(316,)),
        tf.keras.layers.Dense(units=158, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(units=158, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(units=1, activation='linear')
    ])
    regressor.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mean_absolute_error'])
    return regressor
```

```python
regressor = KerasRegressor(model=criar_modelo, epochs=100, batch_size=300)
```

```python
resultados = cross_val_score(estimator=regressor, X=X, y=y, cv=5, scoring='neg_mean_absolute_error')
```

```python
def melhor_modelo(loss):
    regressor = Sequential([
        tf.keras.layers.InputLayer(shape=(315,)),
        tf.keras.layers.Dense(units=158, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(units=158, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(units=1, activation='linear')
    ])
    regressor.compile(optimizer='adam', loss=loss, metrics=['mean_absolute_error'])
    return regressor
```

```python
rede_neural_Tunada= KerasClassifier(model=melhor_modelo)
```

```python
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasRegressor
from tensorflow.keras import backend as k
param_grid = {
    'model__loss': [
        'mean_squared_error',
        'mean_absolute_error',
        'mean_absolute_percentage_error',
        'mean_squared_logarithmic_error',
        'squared_hinge'
    ],
    'epochs': [30],
    'batch_size': [300]
}

```

```python
grid_result = GridSearchCV(estimator=rede_neural_Tunada, param_grid=param_grid, scoring='neg_mean_absolute_error', cv=3)
grid_result = grid_result.fit(X, y)
```

```python
print("Melhor função de custo:", grid_result.best_params_)
print("Melhor score (neg MAE):", grid_result.best_score_)
```
